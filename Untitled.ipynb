# Dropping uid column

```python
inp0.drop("uid", axis=1, inplace=True)
inp0.head()
```

```python
inp0.drop("synopsis",axis=1, inplace=True)
```

```python
inp0.drop("img_url",axis=1, inplace=True)
```

```python
inp0.drop("link",axis=1, inplace=True)
```

```python
inp0[['Starting','Ending']] = inp0['aired'].str.split('to', 1, expand=True)
```

```python
inp0.drop("aired",axis=1, inplace=True)
```

```python
inp0['Starting Year']=inp0['Starting'].str.split(',',1).str[1]
inp0['Ending Year']=inp0['Ending'].str.split(',',1).str[1]
```

```python
inp0.drop("Starting",axis=1, inplace=True)
inp0.drop("Ending",axis=1, inplace=True)
```

```python
inp0['Starting Year']=inp0.astype("int64",errors='ignore')['Starting Year']
```

```python
inp0['Ending Year']=inp0.astype("int64",errors='ignore')['Ending Year']
```
